# Install prereqs for leela-chess

Check to make sure `clinfo` identifies the GPU.

```
%%bash
apt-get update
apt install --fix-missing cmake nvidia-cuda-toolkit git-all libboost-all-dev libopenblas-dev opencl-headers ocl-icd-libopencl1 ocl-icd-opencl-dev zlib1g-dev
apt install clinfo && clinfo


```

# Install and build Leela-chess

```
%%bash
rm -rf leela-chess
git clone https://github.com/glinscott/leela-chess.git
cd leela-chess
git submodule update --init --recursive
mkdir -p build && cd build
cmake ..
make
```

# Install google drive access (optional)

One of the weird things about Google Colab is that each cell spawns a new shell, so if you cd to a directory in one cell, you need to cd to that same directory in another cell.  Also, after about 12 hours, all your files disappear, so you have to rerun everything here. To get around that fact, we save everything to a google drive.



```
%%bash
apt-get install -y -qq software-properties-common python-software-properties module-init-tools
add-apt-repository -y ppa:alessandro-strada/ppa 2>&1 > /dev/null
apt-get update -qq 2>&1 > /dev/null
apt-get -y install -qq google-drive-ocamlfuse fuse

```

# Create run directory to save results in google drive



```
%%bash
mkdir -p drive/LeelaChess
cd drive/LeelaChess
cp /content/leela-chess/build/lczero .
rm -f client_linux
wget -c https://github.com/glinscott/leela-chess/releases/download/v0.6/client_linux
chmod +x client_linux
```

# Run Leela-chess

Don't forget to set your username and password so you can show up on the leaderboards.

If you mounted google drive, then you should be able to see the logfiles generating as you run `client_linux`

```
%%bash
cd drive/LeelaChess
./client_linux --help
```

```
%%bash
USER=<your-leelachess-username>
PASS=<your-leelachess-password>
cd drive/LeelaChess
chmod +x client_linux 
./client_linux -user $USER -password $PASS -gpu 0

```
